```python
import networkx as nx
from networkx.algorithms import centrality
from networkx.readwrite import gexf
from networkx.drawing import layout
import pandas as pd
import matplotlib.pyplot as plt

G = gexf.read_gexf('3mj/subgraph--1-start.gexf')
print(nx.__version__)
print(nx.info(G))
```

```python
ego_node = 'minecraft-java-edition'
radius = 2

G_ego = nx.ego_graph(G, ego_node, radius=radius)
print(nx.info(G_ego))
```

```python
sp = layout.kamada_kawai_layout(G_ego)
```

```python
# from  networkx_viewer import Viewer
```

```python
nx.draw(G_ego, pos=sp, node_color='#0C84B6', node_size=5, with_labels=False )
nx.draw_networkx_nodes(G_ego, pos=sp, nodelist=[ego_node], node_size=10, node_color='#CDE292')
nx.draw_networkx_edges(G_ego, pos=sp, alpha=0.04, edge_color='#D8E7F5')


plt.axis('off')
plt.savefig('images/minecraft_ego_net.png', dpi=300)

```
